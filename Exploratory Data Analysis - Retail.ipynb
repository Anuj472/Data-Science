## Scatterplot

## Stats vs Profit barplot

```python
sales_profit = df.groupby("State")["Profit"].sum()
sales_profit
```

```python
plt.figure(figsize =(17,12))
sns.barplot(x=sales_profit.index,y=sales_profit.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

## Category vs Profit barplot

```python
sales_profit1=df.groupby("Category")['Profit'].sum()
sales_profit1
```

```python
plt.figure(figsize=(17,12))
sns.barplot(x=sales_profit1.index,y=sales_profit1.values)
plt.ylabel("Profit")
plt.xticks(rotation=90)
plt.show()
```

## Sub-category vs Region barplots

```python
plt.figure(figsize=(20,10))
sns.countplot(x='Sub-Category',hue = "Region",data = df)
plt.xticks(rotation = 'vertical')
plt.plot()
```

Conclusion:
* 1.Binders are the mainly sold products followed by paper and Furnishings.\n
* 2.Supplies,Copiers and Machines are the least sold products.\n
* 3.The West region always seems to load in sales compare to other regions.

## Sales per State plots

```python
sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False)
sales.plot.bar(figsize=(30,15))
```

## Profit and Sales for Sub-Categories

```python
category = df.groupby("Sub-Category")["Profit","Sales"].sum()
category.plot.bar(figsize=(20,10))
```

Conclusion:
* We can see that although our sales is high as compared to our profit.
* The best vs Profit ratio is obtaining from copiers and paper.
* When it comes to Tables Bookcase and Supplies as compared to our sales we are making sales.
